```python
#DFS code

def dfs(graph, start, visited=None):
    if visited is None:
        visited = set()
    visited.add(start)
    print(start, end=" ")

    for neighbor in graph[start]:
        if neighbor not in visited:
            dfs(graph, neighbor, visited)

# Example usage
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}
dfs(graph, 'A')  # Output: A B D E F C
```

```python
#BFS code to find Goal node

from collections import deque

# Example graph as an adjacency list
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['G'],
    'F': [],
    'G': []
}

def bfs(graph, start, goal):
    queue = deque([start])  # Queue for BFS
    visited = set()  # Set to track visited nodes

    while queue:
        node = queue.popleft()  # Dequeue from the front
        if node not in visited:
            visited.add(node)
            print(f"Visiting: {node}")
            if node == goal:
                print("Goal found!")
                return True

            # Add neighbors to the queue
            queue.extend(graph[node])

    print("Goal not found.")
    return False

# Testing BFS
start_node = 'A'
goal_node = 'G'

print("BFS Search:")
bfs(graph, start_node, goal_node)
```

```python
# Example graph as an adjacency list
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['G'],
    'F': [],
    'G': []
}

def dfs(graph, start, goal):
    stack = [start]  # Stack for DFS
    visited = set()  # Set to track visited nodes

    while stack:
        node = stack.pop()  # Pop from the stack
        if node not in visited:
            visited.add(node)
            print(f"Visiting: {node}")
            if node == goal:
                print("Goal found!")
                return True

            # Add neighbors to the stack
            stack.extend(reversed(graph[node]))  # Reverse to maintain order

    print("Goal not found.")
    return False

# Testing DFS
start_node = 'A'
goal_node = 'G'

print("DFS Search:")
dfs(graph, start_node, goal_node)
```

```python
#8 puzzle game
import tkinter as tk
import random
from tkinter import messagebox

class Puzzle8:
    def __init__(self, root):
        self.root = root
        self.root.title("8 Puzzle Game")
        self.tiles = []
        self.empty_tile = (2, 2)  # Empty tile position

        self.frame = tk.Frame(self.root)
        self.frame.grid()

        self.create_board()
        self.shuffle_board()
        self.root.bind("<Key>", self.move_tile)

    def create_board(self):
        """Create 3x3 grid with buttons."""
        self.tiles = [[tk.Button(self.frame, text="", font=("Arial", 20), width=5, height=2) for _ in range(3)] for _ in range(3)]
        for i in range(3):
            for j in range(3):
                self.tiles[i][j].grid(row=i, column=j, padx=2, pady=2)
        self.tiles[2][2].config(text="", bg="gray")  # Empty space

    def shuffle_board(self):
        """Shuffle tiles while ensuring solvability."""
        numbers = list(range(1, 9)) + [None]
        while True:
            random.shuffle(numbers)
            if self.is_solvable(numbers):
                break

        idx = 0
        for i in range(3):
            for j in range(3):
                if numbers[idx] is None:
                    self.empty_tile = (i, j)
                    self.tiles[i][j].config(text="", bg="gray")
                else:
                    self.tiles[i][j].config(text=str(numbers[idx]), bg="SystemButtonFace")
                idx += 1

    def is_solvable(self, tiles):
        """Check if a puzzle configuration is solvable."""
        flat_list = [num for num in tiles if num is not None]
        inversions = sum(1 for i in range(len(flat_list)) for j in range(i + 1, len(flat_list)) if flat_list[i] > flat_list[j])
        return inversions % 2 == 0

    def move_tile(self, event):
        """Move tiles using arrow keys."""
        i, j = self.empty_tile
        if event.keysym == "Up" and i < 2:
            self.swap(i, j, i + 1, j)
        elif event.keysym == "Down" and i > 0:
            self.swap(i, j, i - 1, j)
        elif event.keysym == "Left" and j < 2:
            self.swap(i, j, i, j + 1)
        elif event.keysym == "Right" and j > 0:
            self.swap(i, j, i, j - 1)

        if self.is_solved():
            messagebox.showinfo("Congratulations!", "You solved the puzzle!")
            self.shuffle_board()  # Restart after win

    def swap(self, i1, j1, i2, j2):
        """Swap the empty tile with a neighbor tile."""
        self.tiles[i1][j1].config(text=self.tiles[i2][j2]['text'], bg="SystemButtonFace")
        self.tiles[i2][j2].config(text="", bg="gray")
        self.empty_tile = (i2, j2)

    def is_solved(self):
        """Check if the puzzle is solved."""
        correct_numbers = [str(i) for i in range(1, 9)] + [""]
        current_numbers = [self.tiles[i][j]['text'] for i in range(3) for j in range(3)]
        return current_numbers == correct_numbers

if __name__ == "__main__":
    root = tk.Tk()
    Puzzle8(root)
    root.mainloop()
```

```python
#tic tac toe using GUI

import tkinter as tk
from tkinter import messagebox

root = tk.Tk()
root.title("Tic-Tac-Toe")

current_player = "X"
buttons = []
game_active = True

def check_winner():
    global game_active
    winning_combinations = [
        (0, 1, 2), (3, 4, 5), (6, 7, 8),
        (0, 3, 6), (1, 4, 7), (2, 5, 8),
        (0, 4, 8), (2, 4, 6)
    ]
    for combo in winning_combinations:
        if buttons[combo[0]]["text"] == buttons[combo[1]]["text"] == buttons[combo[2]]["text"] != "":
            for index in combo:
                buttons[index].config(bg="green")
            messagebox.showinfo("Tic-Tac-Toe", f"Player {buttons[combo[0]]['text']} wins!")
            game_active = False
            return
    if all(button["text"] != "" for button in buttons):
        messagebox.showinfo("Tic-Tac-Toe", "It's a tie!")
        game_active = False

def button_click(index):
    global current_player
    if buttons[index]["text"] == "" and game_active:
        buttons[index]["text"] = current_player
        check_winner()
        if game_active:
            current_player = "O" if current_player == "X" else "X"

def reset_game():
    global current_player, game_active
    current_player = "X"
    game_active = True
    for button in buttons:
        button.config(text="", bg="SystemButtonFace")

frame = tk.Frame(root)
frame.pack()

for i in range(9):
    button = tk.Button(frame, text="", font=('normal', 40), width=5, height=2,
                       command=lambda i=i: button_click(i))
    button.grid(row=i//3, column=i%3)
    buttons.append(button)

reset_button = tk.Button(root, text="Restart Game", font=('normal', 20), command=reset_game)
reset_button.pack(pady=20)

root.mainloop()
```

```python
#water jug problem

from collections import deque

def water_jug_bfs(capacity_a, capacity_b, target):
    queue = deque([(0, 0, [])])
    visited = set()

    while queue:
        a, b, path = queue.popleft()
        if a == target or b == target:
            return path + [(a, b)]
        visited.add((a, b))

        # Fill jug A
        if (capacity_a, b) not in visited:
            queue.append((capacity_a, b, path + [(a, b)]))
        # Fill jug B
        if (a, capacity_b) not in visited:
            queue.append((a, capacity_b, path + [(a, b)]))
        # Empty jug A
        if (0, b) not in visited:
            queue.append((0, b, path + [(a, b)]))
        # Empty jug B
        if (a, 0) not in visited:
            queue.append((a, 0, path + [(a, b)]))
        # Pour from A to B
        pour = min(a, capacity_b - b)
        if (a - pour, b + pour) not in visited:
            queue.append((a - pour, b + pour, path + [(a, b)]))
        # Pour from B to A
        pour = min(b, capacity_a - a)
        if (a + pour, b - pour) not in visited:
            queue.append((a + pour, b - pour, path + [(a, b)]))

# Example usage
print(water_jug_bfs(4, 3, 2))
```

```python
#travelling Salesman problem


from itertools import permutations

def tsp(graph, start):
    nodes = list(graph.keys())
    nodes.remove(start)
    min_cost = float('inf')
    best_path = None

    for perm in permutations(nodes):
        cost = 0
        current = start
        for node in perm:
            cost += graph[current][node]
            current = node
        cost += graph[current][start]
        if cost < min_cost:
            min_cost = cost
            best_path = (start,) + perm + (start,)

    return best_path, min_cost

# Example usage
graph = {
    'A': {'B': 10, 'C': 15, 'D': 20},
    'B': {'A': 10, 'C': 35, 'D': 25},
    'C': {'A': 15, 'B': 35, 'D': 30},
    'D': {'A': 20, 'B': 25, 'C': 30}
}
print(tsp(graph, 'B'))
```

```python
#tower of hanoi
def tower_of_hanoi(n, source, target, auxiliary):
    if n == 1:
        print(f"Move disk 1 from {source} to {target}")
        return
    tower_of_hanoi(n - 1, source, auxiliary, target)
    print(f"Move disk {n} from {source} to {target}")
    tower_of_hanoi(n - 1, auxiliary, target, source)

# Example usage
tower_of_hanoi(3, 'A', 'C', 'B')
```

```python
class Node:
    def __init__(self, name, children=None, value=None):
        self.name = name
        self.children = children if children is not None else []
        self.value = value

def evaluate(node):
    return node.value

def is_terminal(node):
    return node.value is not None

def get_children(node):
    return node.children

def alpha_beta_pruning(node, depth, alpha, beta, maximizing_player):
    if depth == 0 or is_terminal(node):
        return evaluate(node)

    if maximizing_player:
        max_eval = float('-inf')
        for child in get_children(node):
            eval = alpha_beta_pruning(child, depth-1, alpha, beta, False)
            max_eval = max(max_eval, eval)
            alpha = max(alpha, eval)
            if beta <= alpha:
                break  # Beta cut-off
        return max_eval
    else:
        min_eval = float('inf')
        for child in get_children(node):
            eval = alpha_beta_pruning(child, depth-1, alpha, beta, True)
            min_eval = min(min_eval, eval)
            beta = min(beta, eval)
            if beta <= alpha:
                break  # Alpha cut-off
        return min_eval

# Create the game tree
D = Node('D', value=3)
E = Node('E', value=5)
F = Node('F', value=6)
G = Node('G', value=9)
H = Node('H', value=1)
I = Node('I', value=2)

B = Node('B', children=[D, E, F])
C = Node('C', children=[G, H, I])

A = Node('A', children=[B, C])

# Run the alpha-beta pruning algorithm
maximizing_player = True
initial_alpha = float('-inf')
initial_beta = float('inf')
depth = 3  # Maximum depth of the tree

optimal_value = alpha_beta_pruning(A, depth, initial_alpha, initial_beta, maximizing_player)
print(f"The optimal value is: {optimal_value}")
```

```python
import tkinter as tk
from tkinter import messagebox

class EightQueens:
    def __init__(self, root):
        self.root = root
        self.root.title("8-Queens Game")
        self.board = [[0 for _ in range(8)] for _ in range(8)]
        self.buttons = [[None for _ in range(8)] for _ in range(8)]
        self.queens = 0  # Track number of queens placed
        self.create_board()

    def create_board(self):
        for i in range(8):
            for j in range(8):
                color = "white" if (i + j) % 2 == 0 else "black"
                button = tk.Button(
                    self.root, text="", bg=color, width=4, height=2,
                    command=lambda i=i, j=j: self.place_queen(i, j)
                )
                button.grid(row=i, column=j)
                self.buttons[i][j] = button

    def place_queen(self, row, col):
        if self.board[row][col] == 0:  # Placing a queen
            if self.is_valid(row, col):
                self.board[row][col] = 1
                self.queens += 1
                self.buttons[row][col].config(text="Q", fg="red")
                if self.queens == 8:
                    messagebox.showinfo("Congratulations!", "You solved the 8-Queens problem!")
            else:
                messagebox.showwarning("Invalid Move", "Queens are attacking each other!")
        else:  # Removing a queen
            self.board[row][col] = 0
            self.queens -= 1
            self.buttons[row][col].config(text="")

    def is_valid(self, row, col):
        # Check row
        if sum(self.board[row]) > 0:
            return False

        # Check column
        for i in range(8):
            if self.board[i][col] == 1:
                return False

        # Check diagonals
        for i in range(8):
            for j in range(8):
                if self.board[i][j] == 1:
                    if abs(row - i) == abs(col - j):
                        return False

        return True

# Main function
if __name__ == "__main__":
    root = tk.Tk()
    game = EightQueens(root)
    root.mainloop()
```
